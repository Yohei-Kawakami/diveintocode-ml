```python
"""

136地点から０地点に向かって降る勾配を求める

1.勾配を求める関数を宣言
2.136地点から０地点に向かう配列を作成
3.各地点ごとの勾配の配列を作成

"""

def calculate_grad (x=136):

   grad_mt = np.array([])
   listx = []

   for num0 in range(136):

       num0_true = 135 - num0
       listx.append(num0_true)


   grad_mt = np.array([])

   for num in listx:

       if num > 0:
           grad_mt = np.append(grad_mt, (fuji[num][3]- fuji[num-1][3])
                                     / (fuji[num][0] - fuji[num-1][0]))
       else:
           break

   return grad_mt

grad_mt = calculate_grad(x=136)

print(grad_mt)
len(grad_mt)
```

```python

def down_mt_fuji():
    alpha = 0.15
    number_of_trial = 100

    now_point = 136
    move_point = now_point
    
    now_grad = fuji[now_point][3] - fuji[now_point - 1][3] 

    down_list_grad = []
    down_list_elevation = []
    
    
    
    for _ in range(number_of_trial):
        para_grad = int(round(alpha*now_grad))
        move_point -= para_grad
        now_grad = (fuji[now_point][3] - fuji[move_point][3])/(fuji[now_point][0] - fuji[move_point][0] + (1e-10))
        
        down_list_grad.append(now_grad)
        down_list_elevation.append(fuji[move_point][3])
        now_point = move_point
        
    print(down_list_grad)
    print(down_list_elevation)
        
        
        
down_mt_fuji()
        


    

```
