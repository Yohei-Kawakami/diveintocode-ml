```python
from statistics import mean, variance
from scipy import stats

data = []
for i in range(len(Y)):
    data.append(Y[i,0])

# 標本平均を計算
mean = mean(data)

# 標本分散を計算
variance = variance(data)

# 異常度を計算
scores = []
for x in data:
    score = (x - mean)**2 / variance
    scores.append(score)

# カイ二乗分布による1%水準の閾値
threshold = stats.chi2.interval(0.99, 1)[1]

# 結果の描画
plt.plot(scores, "x", color = "g")
plt.plot([0,365],[threshold, threshold], 'k-', color = "b", ls = "dashed")
plt.xlabel("Sample number")
plt.ylabel("Anomaly score")
plt.ylim([0,10])
plt.show()
```
